```python
#Data Visualization
# Distribution of a numerical variable
plt.figure(figsize=(8, 6))
sns.histplot(Auto['price'], bins=20, kde=True)
plt.title('Distribution of Price')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
Auto
```

```python
# Relationship between body style  and Price
plt.figure(figsize=(10, 6))
sns.boxplot(x='body-style', y='price', data= Auto)
plt.title('Price by Body Style')
plt.xlabel('Body Style')
plt.ylabel('Price')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

```python
#Analysis: Convertible body style are higher in price than hatchback
```

```python
#Relationship between engine size and price, the data points by engine type using different colors

plt.figure(figsize=(10, 6))
sns.scatterplot(x='engine-size', y='price', data=Auto, hue='engine-type')
plt.title('Engine Size vs Price')
plt.xlabel('Engine Size')
plt.ylabel('Price')
plt.legend(title='Engine Type')
plt.show()
```

```python
# Analysis:larger is the engine size , higher is the prices due to increased power and performance
```

```python
# Relationship between two numerical variable ('engine-size' and 'horsepower')
sns.scatterplot(x='engine-size', y='horsepower', data=Auto)
plt.title('Engine Size vs. Horsepower')
plt.xlabel('Engine Size')
plt.ylabel('Horsepower')
plt.show()
```

```python
#Analysis: This indicates that as the engine size increases, the horsepower tends to increase as well.Performance Segmentation: Analyzing the relationship between horsepower and engine size can help in segmenting vehicles based on their performance characteristics. For example, vehicles with larger engine sizes and higher horsepower values might be categorized as high-performance vehicles, while those with smaller engine sizes and lower horsepower values might be considered more fuel-efficient or economical options.
```

```python
#  Relationship of symboling and engine type:
plt.figure(figsize=(10, 6))
sns.countplot(x='engine-type', hue='symboling', data = Auto)
plt.title('Symboling Distribution by Engine Type')
plt.xlabel('Engine Type')
plt.ylabel('Count')
plt.legend(title='Symboling')
plt.show()
```

```python
#Analysis: ohc engine type are more riskier
```

```python
#Relationship between Horse Power vs Price

plt.figure(figsize=(8, 4))
sns.regplot(x='horsepower', y='price', data=Auto)
plt.title('Horsepower vs Price')
plt.xlabel('Horsepower')
plt.ylabel('Price')
plt.show()
```

```python
#Analysis : : The regression line has a positive slope, indicating that as horsepower increases, the price of the automobile tends to increase as well. This suggests that there is a positive correlation between horsepower and price
#Analysis : : The regression line has a positive slope, indicating that as horsepower increases, the price of the automobile tends to increase as well. This suggests that there is a positive correlation between horsepower and price
```
